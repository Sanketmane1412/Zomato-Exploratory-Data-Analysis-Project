```python
# Loading 2nd datasets
code=pd.read_excel("Country-Code.xlsx")
code.head()
```

```python
# Merging both datasets 

comb_data=pd.merge(zomato,code,on="Country Code")
comb_data.head()
```

```python
comb_data.columns
```

```python
# Pie chart to analyze the top 5 countries transactions to avoid overlapping of data
country_names=comb_data.Country.value_counts().values
country_labels=comb_data.Country.value_counts().index
plt.pie(country_names[:5],labels=country_labels[:5],autopct="%1.1f%%")
plt.show()
# Observation- 1)Maximum number of transactions are from india is 90.6%.
# 2)Then united states with 4.5%
# 3)Then united kingdom with 0.8%
```

```python
# More Analysis on Rating
# Lets groupby all rating features from the datasets
rating=comb_data.groupby(['Aggregate rating', 'Rating color', 'Rating text','Votes']).size().reset_index().rename(columns={0:"Rating count"},inplace=False)
rating.head()
```

```python
# Observation -
# 1) rating from 4.5 - 4.9 is excellent and indicated with dark green colour
# 2) rating from 4.0 - 4.4 is very good and indicated with green colour
# 3) rating from 3.5 - 3.9 is good and indicated with orange colour
# 4) rating from 3.0 - 3.4 is poor and indicated with yellow colour
# more till white
```

```python
# Bar plot to analyze the relationship between aggregate and rating color
plt.rcParams['figure.figsize']=(12,6)
ax=sns.barplot(data=rating,x="Aggregate rating",y="Rating count",palette="Set2")
for x in ax.containers:
    ax.bar_label(x)
plt.show()
# Observation
# 1)Not rated count is high
# 2)max number of rating is between 2.5 to 3.5
```

```python
# Analyzing counting number the Aggregate rating based on rating colours
plt.rcParams['figure.figsize']=(12,6)
ax=sns.countplot(hue="Rating color",x="Aggregate rating",data=rating,palette=["white","red","orange","yellow","green","darkgreen"])
for x in ax.containers:
    ax.bar_label(x)
plt.show()
#Observation- Aggreate rating is high between 3.4 to 4.2 with count of between 194 to 278 max.
```

```python
# Analyzing or counting the total rating counts based on rating colours
plt.rcParams['figure.figsize']=(12,6)
ax=sns.countplot(x="Rating color",data=rating,palette=["white","red","orange","yellow","green","darkgreen"])
for x in ax.containers:
    ax.bar_label(x)
plt.show()
# Observation - 
# Rating colors here white color has total 4 customers rated
# Rating colors here red color has total 157 customers rated
# Rating colors here orange color has total 1052 customers rated
# Rating colors here yellow color has total 1268 customers rated
# Rating colors here green color has total 944 customers rated
# Rating colors here dark green color has total 288 customers rated
```

```python
def plot(country_code):
    country_df = comb_data[comb_data["Country Code"] == country_code]

    avg_rating = (country_df.groupby("City")["Aggregate rating"]
                  .mean()
                  .sort_values(ascending=True)
                  .head(15))
    
    plt.figure(figsize=(12,6))
    avg_rating.plot(kind='bar', color='skyblue')
    plt.title(f"Top 15 Cities by Average Rating (Country Code: {country_code})")
    plt.xlabel("City")
    plt.ylabel("Average Aggregate Rating")
    plt.xticks(rotation=45, ha='right')
    plt.show()

plot(162)  
# Observation - Created a function that creates a bar graph when we pass specific country name 
# here in phillippines city Quezon has highest aggregate rating
```

```python
# Find top 10 cuisines accoring to countries using pie chart by creating function
# Here we can change country it will automatically analyze 10 countries in [:--]
def plot_pie(Country):
    cuisines_name=comb_data[comb_data["Country"]==Country].Cuisines.value_counts().values
    cuisines_label=comb_data[comb_data["Country"]==Country].Cuisines.value_counts().index
    plt.pie(cuisines_name[:10],labels=cuisines_label[:10],autopct="%1.2f%%",shadow=True)
    plt.title("Cuisines of Uae")
    plt.show()
plot_pie("UAE")
```

```python
plt.rcParams['figure.figsize']=(12,6)
col=["Aggregate rating","Votes","Price range"]
graph=sns.pairplot(data=comb_data[col],palette="Set2")
graph.map_upper(sns.scatterplot)
graph.map_diag(sns.barplot)
graph.map_lower(sns.histplot)
graph.add_legend()
plt.show()
```
